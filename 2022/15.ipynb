```python
def get_range(lst):
    m0 = 0
    m1 = 0
    lst_new = []
    for i, j in lst:
        i = max(i, 0)
        j = min(j, 4000000)
        
        if m1 >= j:
            continue
            
        elif (m1 <= j) & (m1 >= i):
            m1 = max(m1, j)
        else:
            lst_new.append([m0, m1])
            m0 = i
            m1 = j
            print('renew:', (i, j))
            
    lst_new.append([m0, m1])
    return lst_new
        
```
