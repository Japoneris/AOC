```python
with open("input/07", "r") as fp:
    data = fp.read().split('\n')[:-1]
```

```python
path = "/"

dico = {} # Path: {"next":[], "size": 0}

cnt = 0
lst = []

for row in data:
    if row.startswith("$ cd "):
        px = row.split(" ")[2]
        if px == "/":
            path = "/"
            
        elif px == "..":
            path = path.rsplit("/", 2)[0] + "/"
        else:
            path = path + px + "/"
            
            
    elif row.startswith("$ ls"):
        print(path)
        dico[path] = {"next": [], "size": 0}
        
    else:
        print("\t", row)
        
        if "dir" in row:
            a, b = row.split(" ", 1)
            dico[path]["next"].append(path + b + "/")
            
        else:
            a, b = row.split(" ", 1)
            dico[path]["size"] += int(a)
            
            
```

```python
list(filter(lambda x: len(x["next"]) == 0, dico.values()))
```

```python
final_size = {}

lst_cands = list(dico)

while len(lst_cands) > 0:
    print("Left:", len(lst_cands))
    
    a = lst_cands.pop(0)
    lx = dico[a]["next"] 
    if len(set(lx) - set(final_size)) == 0:
        # Compute the score
        cnt = dico[a]["size"]
        for xx in lx:
            cnt += final_size[xx]
        
        final_size[a] = cnt
        
    else:
        # Nothing can be done, add it to the next
        lst_cands.append(a)
    
```

```python
limit = 100000


# At most #limit
print(sum(filter(lambda x: x <= limit, final_size.values())))
```
