The starting point, the lava pool, is the top-left city block; the destination, the machine parts factory, is the bottom-right city block. (Because you already start in the top-left block, you don't incur that block's heat loss unless you leave that block and then return to it.)

ecause it is difficult to keep the top-heavy crucible going in a straight line for very long, it can move **at most three blocks in a single direction before it must turn 90 degrees** left or right.

# Part 1

```python
def check_valid(x, y, n0, n1):
    if (x < 0) | (y < 0) | (x >= n0) | (y >= n1):
        return False
    else:
        return True
    
def update_dir(x, y, d):
    if d == "R":
        return (x, y+1)
    elif d == "L":
        return (x, y-1)
    elif d == "U":
        return (x-1, y)
    elif d == "D":
        return (x+1, y)
```

```python
def simultaneous_exploration(M):
    
    # x, y, dir, move, heatloss
    start = (0, 0, "D", 0, 0)
    lst = [start]
    
    n0, n1 = M.shape
    mmax = M.sum()
    B = np.ones((n0, n1)) * mmax
    B[0, 0] = 0
    
    
    dic_idd = {"R": 0, "L": 1, "U":2, "D": 3}
    
    cnt = 0
    while len(lst) != 0:
        
            
        
        cnt += 1
        if cnt % 1000 == 0:
            
            
            # Filter all elements in the same cell
            dic_new = {}
            for x, y, d, m, h in lst:
                key = "{}_{}_{}_{}".format(x, y, d, m)
                if key in dic_new:
                    dic_new[key] = min(dic_new[key], h)
                else:
                    dic_new[key] = h
            
            lst_new = []
            for key, h in dic_new.items():
                x, y, d, m = key.split("_")
                x, y, m = int(x), int(y), int(m)
                
                b_best = B[x, y]
                if h < b_best:
                    B[x, y] = h
                    
                elif h > b_best + 30:
                    # No need to explore, too costly
                    continue
                    
                lst_new.append((x, y, d, m, h))
            
            
            # Useful to be faster
            # Lowest heat first
            lst = sorted(lst_new, key=lambda x: x[4])
            tot = (B != mmax).sum()
            print("{} / {} explored".format(tot, n0*n1), end="\t")
        
        x, y, d, m, h = lst.pop(0)
        if cnt % 1000 == 0:
            print("{} ({}, {} : {})|".format(len(lst), x, y, h))
            
        
        # easy stopping
        if h > B[-1, -1]:
            continue
        
        b_best = B[x, y]
        if h < b_best:
            B[x, y] = h
            
        elif h > b_best + 30:
            # No need to explore, too costly
            continue
            
        
        if m < 3:
            # Single possibility
            x1, y1 = update_dir(x, y, d)
            if check_valid(x1, y1, n0, n1):
                lst.append((x1, y1, d, m+1, h + M[x1, y1]))
            
            # Otherwise, do not save
            
        if m > 0 :
            # Should turn, no heat update
            if d in "RL":
                lst.append((x, y, "U", 0, h))
                lst.append((x, y, "D", 0, h))
                
            elif d in "UD":
                lst.append((x, y, "R", 0, h))
                lst.append((x, y, "L", 0, h))
        
        
    return B
                
    
```
