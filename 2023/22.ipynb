#### Process input

```python
def process_line(line):
    a, b = line.split("~")
    C1 = list(map(int, a.split(",")))
    C2 = list(map(int, b.split(",")))
    return C1, C2

def process_puzzle(ttt):
    lines = ttt.split('\n')
    parts = list(map(process_line, lines))
    X1 = np.array(list(map(lambda x: x[0], parts)))
    X2 = np.array(list(map(lambda x: x[1], parts)))
    return X1, X2
```

##### Print / debug

```python
def print_matrix(X1, X2):
    
    mz = 0
    Mz = max(X1[:, 1].max(), X2[:, 1].max()) + 1
    
    mx = 0
    Mx = max(X1[:, 0].max(), X2[:, 0].max()) + 1
    
    M_tot = np.zeros((Mz - mz, Mx - mx))
    for idx, ((x0, z0), (x1, z1)) in enumerate(zip(X1, X2)):
        M_tot[z0:z1+1, x0:x1+1] = idx + 1
        
    return M_tot[::-1]
    

def print_over_xz(X1, X2):
    
    X1 = X1[:, [0, 2]]
    X2 = X2[:, [0, 2]]
    
    return print_matrix(X1, X2)
    

def print_over_yz(X1, X2):
    
    X1 = X1[:, [1, 2]]
    X2 = X2[:, [1, 2]]
    
    return print_matrix(X1, X2)
    
```

```python
def create_tensor(X1, X2):
    
    a, b, c = X2.max(axis=0)
    
    M_tot = np.zeros((a+1, b+1, c+1))
    
    for idx, ((x0, y0, z0), (x1, y1, z1)) in enumerate(zip(X1, X2)):
        M_tot[z0:z1+1, x0:x1+1, y0:y1+1] = idx + 1
        
    return M_tot
```

### Sort bricks by z

```python
def check_coordinates(a0, a1, b0, b1):
    """
    A #######a1
    B          b0#####
    
    A         a0####a1
    B b0####b1
    
    A         a0####a1
    B     b0####b1
    
    A    a0####a1
    B         b0####b1
    
    A        a0####a1
    B     b0##########b1
    
    
    
    
    """
    
    if (a1 < b0) | (b1 < a0):
        #if (a1 + 1 < b0) | (b1+1 < a0):
        return False
    
    else:
        return True
    
    

def is_connected(x1a, x1b, x2a, x2b):
    """
    xa: brick above
    xb: brick bellow
    """
    
    if x1a[2] == x2b[2] + 1: # Si Z+ touche Z-, connection possible
        # Il y a possiblement un contact si l'une des coordinates x / y se recouvrent
        T1 = check_coordinates(x1a[0], x1b[0], x2a[0], x2b[0])
        T2 = check_coordinates(x1a[1], x1b[1], x2a[1], x2b[1])
        # Si les deux coordinnÃ©es se recouvrent, alors OK
        return T1 & T2
    
    else:
        return False
    
    
```

```python
def eval_brick_down(X1r, X2r, idx):
    # Remove the brick
    X1 = np.concatenate([X1r[:idx], X1r[idx+1:]]).copy()
    X2 = np.concatenate([X2r[:idx], X2r[idx+1:]]).copy()

    
    cnt = 0
    for idx, (x1a, x2a) in enumerate(zip(X1, X2)):
        FLAG_MOVE = False
        FLAG = False
        while FLAG == False:
            if x1a[2] == 1:
                break

            msk = X2[:, 2] + 1 == x1a[2]
            for x1b, x2b in zip(X1[msk], X2[msk]):
                FLAG = FLAG | is_connected(x1a, x2a, x1b, x2b)

            if FLAG == False:
                # Modify X1 / X2 in place. x1a are reference to it
                x1a[2] = x1a[2] - 1
                x2a[2] = x2a[2] - 1
                FLAG_MOVE = True

            else:
                break
                
        if FLAG_MOVE:
            cnt += 1
            
    return cnt
```

```python
tot = 0
for idx in range(len(X1)):
    v = eval_brick_down(X1, X2, idx)
    tot += v
```
