```python
def process_input(ttt):
    lines = ttt.split("\n")
    lst = []
    for line in lines:
        a, b = line.split(": ")
        bbb = b.split(" ")
        for b in bbb:
            lst.append((a, b))
            lst.append((b, a))
           
    S = set(map(lambda x: x[0], lst))
    dic = dict([(c, []) for c in S])
    for a, b in lst:
        dic[a].append(b)
    
    return dic, lst[::2]
```

# Part 1: three wires to disconnect to obtain two groups

```python
def get_distance(a, puzzle):
    dic = {a: 0}

    S_seen = set()
    S = {a}
    d = 1
    while len(S) > 0:
        S_seen = S_seen | S
        lll = []
        for s in S:
            lll.extend(puzzle[s])

        S = []
        for s in set(lll):
            if s not in dic:
                dic[s] = d
                S.append(s)

        S = set(S)
        d += 1
        
    return dic
```

```python
def get_distance_up(a, puzzle, S0):
    dic = {a: 0}
    for s in S0:
        dic[s] = -1

    S_seen = S0
    S = {a}
    d = 1
    while len(S) > 0:
        S_seen = S_seen | S
        lll = []
        for s in S:
            lll.extend(puzzle[s])

        S = []
        for s in set(lll):
            if s not in dic:
                dic[s] = d
                S.append(s)

        S = set(S)
        d += 1
        
    return dic
```

```python
puzzle, conns = process_input(ttt)

n = len(puzzle)
names = list(puzzle)
```

```python
a = names[36]
D0 = get_distance(a, puzzle)

lst_out = []
for ni in names:
    v = D0[ni]
    v1 = max([D0[nj] for nj in puzzle[ni]])
    if (v > v1) & (v >= 10):
        lst_out.append(ni)
        print(v)
```

```python
path = []
for ni in lst_out:
    D1 = get_distance(ni, puzzle)
    D_tot = dict([(i, D0[i] + D1[i]) for i in D0])
    D_tot = dict(sorted(D_tot.items(), key=lambda x: x[1]))

    v_min = min(D_tot.values())
    x_min = list(filter(lambda x: D_tot[x] == v_min, D_tot))
    dico = {}
    for k in x_min:
        dico[k] = D1[k]
        
    path.append((ni, dico))
```

##### Decompose path

```python
subpath = []
for ni, dico in path:
    lst = [[ni]]
    d = 0
    while len(lst) != 0:
        subi = lst.pop()
        c = subi[-1]
        if c == a:
            subpath.append((ni, subi))
            continue
            
        else:
            dc = dico[c]
            S = set(puzzle[c]) & set(filter(lambda x: dico[x] == dc+1, dico))
            for si in S:
                lst.append(subi + [si])
        
        
    
```
