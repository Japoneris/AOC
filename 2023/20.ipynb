```python
def process_input(ttt):
    lines = ttt.split("\n")
    lines = list(map(lambda x: x.split(" -> "), lines))
    lst = []
    broadcaster = None
    for a, b  in lines:
        if "broadcaster" == a:
            broadcaster = b.split(", ")            
        else:
            op = a[0]
            st = a[1:]

            lst.append((st, [op, b.split(", ")]))
            
    return broadcaster, dict(lst)
```

```python
broad, puzzle = process_input(txt)
```

### Flip-flop modules (prefix %) 

Are either on or off; 

- they are **initially off**. 

- If a flip-flop module receives a **high pulse**, it is ignored and **nothing happens**. 
- if a flip-flop module receives a **low pulse**, it flips between **on and off**. 

When turning:

- If it **was off**, it **turns on** and sends a **high pulse**. 
- If it **was on**, it **turns off** and sends a **low pulse**.




### Conjunction modules (prefix &) 

remember the type of the most recent pulse received from each of their connected input modules; 

they initially **default to remembering a low pulse** for each input. 

When a pulse is received, the conjunction module first updates its memory for that input. 

Behavior 

- Then, if it **remembers high pulses for all inputs**, it **sends a low pulse**; 
- otherwise, it **sends a high pulse**.


### Broadcast 

There is a single broadcast module (named broadcaster). 

- When it receives a pulse, it sends the **same pulse to all of its destination modules**.

### Single button

Here at Desert Machine Headquarters, there is a module with a single button on it called, aptly, the button module. 

When you push the button, a **single low pulse** is sent directly to the **broadcaster module**.

# Part 1

## Pulse loop

##### Create operator memory

```python
dic_memory = {}
lst_esper = []
for k, (op, _) in puzzle.items():
    if op == "%":
        dic_memory[k] = 0
    
    elif op == "&":
        lst_esper.append(k)

for e in lst_esper:
    dic = {}
    for k, (_, succ) in puzzle.items():
        if e in succ:
            dic[k] = 0
            
    dic_memory[e] = dic
```

## Single

```python
lst = []
for mk in broad:
    lst.append(("low", mk, "start"))

cnt_high = 0
cnt_low = 1

while lst != []:
    w, mk, prev = lst.pop(0)
    print("{} = {} => {}".format(prev, w, mk))

    if w == "high":
        cnt_high += 1
    else:
        cnt_low += 1

    if mk == "output":
        continue 

    op, successors = puzzle[mk]

    if op == "%":
        # Flip flop
        if w == "low":
            dic_memory[mk] = 1 - dic_memory[mk]
            # Otherwise, nothing happens

            mess = "high" if dic_memory[mk] == 1 else "low"
            for suc in successors:
                lst.append((mess, suc, mk))

    elif op == "&":
        # Update the module
        dic_memory[mk][prev] = 1 if w == "high" else 0
        
        l = len(list(filter(lambda x: x == 0, dic_memory[mk].values())))
        
        mess = "low" if l == 0 else "high" #: # all high
        for suc in successors:
            lst.append((mess, suc, mk))

print(cnt_low, cnt_high)
```

## RQ: 

We see that RX is activated by &RM.

Therefore, to get RX a low pulse, RM must have all its connected input high.

There are four registers which activate RM:

- &DH,
- &QD:
- &BB
- &DP

DH, QD, BB, and DP should not have all their inputs high

These registers are connected to a single one:

- DH to DT
- QD to XM
- BB to VT
- DP to BR

To get all first register high, all second registers should be low.
Ie, all input of second registers should be high.


Here,  we isolate for each entry when they would all be high

#### Memo

**Conjunction modules** (prefix &) remember the type of the most recent pulse received from each of their connected input modules; 

- they initially default to remembering a **low pulse for each input**. 
- When a pulse is received, the conjunction module first updates its memory for that input. Then, if it remembers **high pulses for all inputs**, it sends a **low pulse**; 
- otherwise, it sends a **high pulse**.

```python
dic_memory = {}
lst_esper = []
for k, (op, _) in puzzle.items():
    if op == "%":
        dic_memory[k] = 0
    
    elif op == "&":
        lst_esper.append(k)

for e in lst_esper:
    dic = {}
    for k, (_, succ) in puzzle.items():
        if e in succ:
            dic[k] = 0
            
    dic_memory[e] = dic


REG = "dp" # qd, dh, bb, dp
REG1 = list(dic_memory[REG].keys())[0]

Patterns_B = []

for CNT in range(1, 10000):

    # Initialization, push the button
    lst = []
    

    for mk in broad:
        lst.append(("low", mk, "start"))
    
    cnt_high = 0
    cnt_low = 0

    while lst != []:
        w, mk, prev = lst.pop(0)

        if mk == REG1:
            Patterns_B.append((CNT, list(dic_memory[REG1].values())))
    
            if sum(dic_memory[REG1].values()) == len(dic_memory[REG1]):
                print("\t=>{}: {} UP".format(CNT, REG))
            
        if mk == "rx":
            continue
            
            
        op, successors = puzzle[mk]
        
        if op == "%":
            if w == "low":
                dic_memory[mk] = 1 - dic_memory[mk]
                # Otherwise, nothing happens

                mess = "high" if dic_memory[mk] == 1 else "low"
                for suc in successors:
                    lst.append((mess, suc, mk))

        elif op == "&":
            # Update the module
            dic_memory[mk][prev] = 1 if w == "high" else 0
            l = len(list(filter(lambda x: x == 0, dic_memory[mk].values())))
            mess = "low" if l == 0 else "high" #: # all high
            for suc in successors:
                lst.append((mess, suc, mk))
    
       
    if CNT % 10000 == 0:
        print("{}\t{}:{}".format(CNT, cnt_high, cnt_low))
```
