```python
def arr_to_patt(arr):
    lst = []
    for v in arr:
        if v == -1:
            lst.append("#")
        elif v == 0:
            lst.append(".")
        else:
            lst.append("?")
            
    return "".join(lst)
```

```python
def convert_to_array(a):
    arr = np.zeros(len(a))
    for i, v in enumerate(a):
        if v == "#":
            arr[i] = -1
        elif v == "?":
            arr[i] = 1
        # else: 0 for "."

    return arr

def process_text(ttt):
    lines = ttt.split("\n")
    lst_sources = []
    for line in lines:
        a, b = line.split(" ")
        
        b = list(map(int, b.split(",")))
        
        lst_sources.append([convert_to_array(a), b])
        
    return lst_sources

def compare_array(arr0, arr1):

    for a, b in zip(arr0, arr1):
        if (a == 1) | (b == 1):
            continue
        
        if (a != b):
            return False
        
    return True
```

```python
def derive_patterns(arr):
    # Generate all possible replacement solution
    
    lst = [[]]
    l = len(arr)
    for i, v in enumerate(arr):
        if v == 1:
            lst1 = list(map(lambda x: x + [0], lst))
            lst2 = list(map(lambda x: x + [-1], lst))
            lst = lst1 + lst2
        else:
            lst = list(map(lambda x: x + [v], lst))
            
    return np.array(lst)
```

```python
def eval_pattern(arr):
    # Create the list out of the blocks
    buffer = 0
    mem = []
    for i in arr:
        if (i == 0) & (buffer != 0):
            mem.append(buffer)
            buffer = 0
        elif i == -1:
            buffer += 1
            
    if buffer != 0:
        mem.append(buffer)
    
    return mem
```

```python
def match_pattern(lst1, lst2):
    if len(lst1) != len(lst2):
        return False
    for i, j in zip(lst1, lst2):
        if i != j:
            return False
    
    return True
```

```python
puzzle = process_text(my_text)
```

```python
tot = 0
for a, b in  puzzle:

    derivatives = derive_patterns(a)
    t = sum(b)
    derivatives = derivatives[(derivatives == -1).sum(axis=1) == t]

    cnt = 0
    for arri in derivatives:
        pat = eval_pattern(arri)
        if match_pattern(b, pat):
            cnt += 1
    print(cnt)
    tot += cnt
    
print("PArt 1: {}".format(tot))
```

```python
def process_text_part2(ttt):
    lines = ttt.split("\n")
    
    lst_sources = []
    for line in lines:
        a, b = line.split(" ")
        b = list(map(int, b.split(",")))
        
        a1 = "..." + "?".join([a for _ in range(5)]) + "....." # Add the last "End character" to avoid overlap
        b1 = b * 5
        
        lst_sources.append([convert_to_array(a1), b1])
        
    return lst_sources
```

```python
def place_pattern_v2(arr, v):
    """Save location where there is a valid block.
    """
    
    msk = (arr == 0).astype(int) # all "." are set to 1
    # -1: #, 0: .,  +1: ?
    
    lst = []
    for i in range(len(arr) - v):
        if msk[i: i+v].sum() == 0: # If true "# or "?", otherwise, there is at least "."
            
            if (arr[i + v] != -1) & (arr[i - 1] != -1): # Block after and block before should not be "#"
                lst.append((i, i + v))
            
    return lst
    
```
