# Processing

```python
def process_input(ttt):
    lines = ttt.split('\n')
    n0 = len(lines)
    n1 = len(lines[1])
    dic_cvt = {"#": 1, ".": 0, ">": 2, "<": 3, "v": 4, "^": 5}
    M = np.zeros((n0, n1))
    for idx, row in enumerate(lines):
        for jdx, v in enumerate(row):
            M[idx, jdx] = dic_cvt[v]
            
    return M
            
```

```python
def move(x, y, M):
    """
    :param history: previous locations
    """
    lst_options = []
    # Right
    if M[x, y+1] == 0:
        lst_options.append((x, y+1, 1))
    elif M[x, y+1] == 2:
        if M[x, y+2] == 0:
            lst_options.append((x, y+2, 2))
        
     
    if M[x, y-1] == 0:
        lst_options.append((x, y-1, 1))
    elif M[x, y-1] == 3:
        if M[x, y-2] == 0:
            lst_options.append((x, y-2, 2))
        
    if M[x-1, y] == 0:
        lst_options.append((x-1, y, 1))
    elif M[x-1, y] == 5:
        if M[x-2, y] == 0:
            lst_options.append((x-2, y, 2))
        
    if M[x+1, y] == 0:
        lst_options.append((x+1, y, 1))
    elif M[x+1, y] == 4:
        if M[x+2, y] == 0:
            lst_options.append((x+2, y, 2))
        
    
    return lst_options
    
```

# Part 1

```python
my_text = txt_test
my_text = txt

M = process_input(my_text)
l0, l1 = M.shape


M[0, 1] = -1 # Do not go here


max_length = 0
cnt = 1

mem = [[(1, 1, 1)]]
history = [[0, 1, 0]] # Where we were before

while len(mem) != 0:
    last = mem.pop(-1)
    if len(last) == 0:
        # Rollback
        x, y, dx = history.pop(-1)
        cnt -= dx
        M[x, y] = 0
        continue
        
    else:
        #Â Take last position
        x, y, dx = last.pop(0)
        
        # Put leftover in the stack
        mem.append(last)
        
        # Where we are
        history.append((x, y, dx))
        M[x, y] = -1
        cnt += dx
    
    
    if (x == l0 -1) & (y == l1 - 2):
        max_length = max(max_length, cnt)
        #print("STOP: {}".format(cnt))
        mem.append([])
        #fig, ax = plt.subplots()
        #plt.imshow(M == -1)
        continue
        
    else:
        opts = move(x, y, M)
        mem.append(opts)
```
